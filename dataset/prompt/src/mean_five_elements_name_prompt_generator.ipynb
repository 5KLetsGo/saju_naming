```python
def name_split(filename):
    name_list = []
    name_str = ''
    i = 0
    with open(f'./categorized_name_hanja/{filename}.txt', 'r', encoding='utf-8') as f:
        while True:
            i += 1
            line = f.readline()
            name_str += line
            if i == 11:
                name_list.append(name_str)
            elif i == 13:
                name_str = ''
                i = 0
            if not line:
                break
    return name_list

f_woodwood = name_split('female_names_목목')
f_woodfire = name_split('female_names_목화')
f_woodwater = name_split('female_names_목수')
f_woodmetal = name_split('female_names_목금')
f_woodearth = name_split('female_names_목토')
f_firefire = name_split('female_names_화화')
f_firewater = name_split('female_names_화수')
f_firemetal = name_split('female_names_화금')
f_fireearth = name_split('female_names_화토')
f_waterwater = name_split('female_names_수수')
f_watermetal = name_split('female_names_수금')
f_waterearth = name_split('female_names_수토')
f_metalmetal = name_split('female_names_금금')
f_metalearth = name_split('female_names_금토')
f_earthearth = name_split('female_names_토토')

m_woodwood = name_split('male_names_목목')
m_woodfire = name_split('male_names_목화')
m_woodwater = name_split('male_names_목수')
m_woodmetal = name_split('male_names_목금')
m_woodearth = name_split('male_names_목토')
m_firefire = name_split('male_names_화화')
m_firewater = name_split('male_names_화수')
m_firemetal = name_split('male_names_화금')
m_fireearth = name_split('male_names_화토')
m_waterwater = name_split('male_names_수수')
m_watermetal = name_split('male_names_수금')
m_waterearth = name_split('male_names_수토')
m_metalmetal = name_split('male_names_금금')
m_metalearth = name_split('male_names_금토')
m_earthearth = name_split('male_names_토토')

print(len(f_woodwood))
```
